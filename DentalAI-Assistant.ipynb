## Library Import

```python
import os
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import ChatPromptTemplate
from datetime import date, datetime
```

```python
# Get the current Date and Day
current_date_str = datetime.now().strftime("%Y-%m-%d")
current_date = datetime.strptime(current_date_str, "%Y-%m-%d")
today = current_date.strftime("%Y-%m-%d")
current_day = current_date.strftime("%A")

# Set up your OpenAI API key
os.environ["OPENAI_API_KEY"] = "OPENAI-API_KEY"

# Initialize the OpenAI ChatGPT model which we want to use
llm = ChatOpenAI(model_name="gpt-4o-mini")

# Define the prompt for GPT
prompt_template = """
You are an AI assistant of a shah's dental clinic. You are here to provide the basic answer of their question related to dental services.
Your task is to guide the patient through the booking process. Answer should not be more than 50 words also add emojis whereever needed.
Today date is {current_date} and day is {current_day}. Clinic time will be 09:00 AM to 8:00 PM only from Monday to Saturday.
You will have to calculate the day if user simply give the date and make sure that day is not sunday. Also check if the date is lesser than tomorrows date then ask user to select another date.
Make sure to:
0. Make sure that if user send the date and it's sunday then gracefully say no and ask to book for another day.
1. Recall previous conversations.
2. Ask for the basic details of the patient like name, age, gender and ask it one by one and make it more conversional and reply in graceful manner.
3. Confirm the date and time.
4. Confirm the booking with a polite response.
5. Ask if you need any further assistance.

Conversation history:
{history}

Patient: {input}

Assistant:
"""

# Now we'll create a chain with the prompt and the LLM
chain = LLMChain(llm=llm, prompt=ChatPromptTemplate.from_template(prompt_template))

# Chat with user with the older history.
def book_appointment_conversation(user_input, history):

    # Generate a response using the chain
    response = chain.run(input=user_input, history=history,current_date=today,current_day=current_day)

    # Append the user input to the history
    history += f"\nPatient: {user_input}"
    # Add assistant's response in the history
    history += f"\nAssistant: {response}"

    return response, history
```

```python
conversation_history = ""

# Iterate the loop till user ask to exit and end the conversation.
print("Assistant: Hello! üòä Welcome to Shah's Dental Clinic! ü¶∑ How can I assist you today regarding your dental needs?")
while True:
    inp = input('You: ')
    if inp.lower() == "exit":
        break
    else:
        response, conversation_history  = book_appointment_conversation(inp,conversation_history )
        print("Assistant:", response)
print("Thank you for connecting with us...!")
```
